```python
import csv

newss = []
# 創建一個空的列表來存儲數據

# 將動畫名稱和觀看數放入數據列表中
for i in range(len(news)):
    newss.append({'新聞標題': news[i]})

# 指定 CSV 檔案名稱
csv_filename = 'new.csv'

# 開啟 CSV 檔案並寫入數據
with open(csv_filename, 'w', newline='', encoding='utf-8-sig') as csv_file:
    fieldnames = ['新聞標題']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    
    writer.writeheader()

    # 寫入數據
    for data in newss:
        writer.writerow(data)

print(f'數據已保存到 {csv_filename}')
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

news = []
base_url = 'https://www.storm.mg/life-category/s24251/'

# 使用Chrome WebDriver
driver = webdriver.Chrome()

try:
    for page in range(1, 3):  # 1到5的范围
        url = f'{base_url}{page}'
        driver.get(url)

        # 使用BeautifulSoup解析网页内容
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        p_elements = soup.select('.card_title')

        # 遍历每个找到的 .card_title 元素，并将其文本内容添加到列表中
        for p_element in p_elements:
            p = p_element.text.strip()
            print(p)
            news.append(p)
finally:
    driver.quit()  # 关闭浏览器窗口
```
